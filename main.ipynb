```python
import pandas as pd
import numpy as np

df = pd.read_csv('./DataSets/customer_support_tickets.csv')
```

```python
df
```

```python
CustomerGender = df['Customer Gender'].value_counts()
pd.DataFrame.to_csv(CustomerGender, './Results/CustomerGender.csv')
```

```python
ticket_counts_per_product = df.groupby('Product Purchased')['Ticket ID'].count()
sorted_ticket_counts = ticket_counts_per_product.sort_values(ascending=False)
print(sorted_ticket_counts)
pd.DataFrame.to_csv(sorted_ticket_counts, './Results/ticket_counts_per_product.csv')
```

```python
# Group data by 'Customer Name' or 'Customer Email' and count the number of tickets for each customer
ticket_counts_per_customer = df.groupby('Customer Name')['Ticket ID'].count()  # Use 'Customer Email' if names can be duplicated

# Sort the customers by the number of tickets in descending order to get the customers with the highest number of issues
sorted_ticket_counts = ticket_counts_per_customer.sort_values(ascending=False)

# Print the sorted ticket counts to see customers with the highest number of issues
print(sorted_ticket_counts)
pd.DataFrame.to_csv(sorted_ticket_counts, './Results/ticket_counts_per_customer.csv')
```

```python
# Group data by 'Product Purchased' and 'Customer Gender', then count the number of tickets
issues_by_product_and_gender = df.groupby(['Customer Gender','Product Purchased'])['Ticket ID'].count()

# Sort the counts in descending order to identify the highest number of issues
sorted_issues = issues_by_product_and_gender.sort_values(ascending=False)

pd.DataFrame.to_csv(sorted_issues,'./Results/sorted_issues.csv')
# Print the sorted list
print(sorted_issues)
```

```python
issues_by_gender = df.groupby(['Customer Gender'])['Ticket ID'].count()
print(issues_by_gender)
pd.DataFrame.to_csv(issues_by_gender,'./Results/issues_by_gender.csv')
```
